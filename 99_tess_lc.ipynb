```python
import matplotlib.pyplot as plt
import scipy
import numpy as np
from astropy.coordinates import SkyCoord
from astropy.io import fits
import lightkurve as lk
import eleanor
from astropy import wcs

plt.rcParams.update({'font.size': 18})
```

```python
f = fits.open('tess/tess-s0015-1-4_324.612020_25.372677_90x90_astrocut.fits')
```

```python
cutout_img = f[1].data[600]['FLUX']
```

```python
cutout_wcs = wcs.WCS(f[2].header,relax=True)
```

```python
fig, ax = plt.subplots(subplot_kw={'projection':cutout_wcs})
ax.imshow(cutout_img, cmap='viridis',vmin=0,vmax=500)
fig.set_size_inches(10,10)

xcoords = ax.coords[0]
ycoords = ax.coords[1]
xcoords.set_major_formatter('d.ddd')
ycoords.set_major_formatter('d.ddd')
xcoords.set_axislabel("RA (deg)")
ycoords.set_axislabel("Dec (deg)")
```
